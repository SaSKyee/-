利用四分位距检测异常值的函数

```python
def erro(x):
    Percentile = np.percentile(x,[0,25,50,75,100])
    IQR = Percentile[3] - Percentile[1]
    Uplimit = Percentile[3] + IQR*1.5
    Downlimit = Percentile[1] - IQR*1.5
    if x.values.all() > Downlimit or x.values.all() < Uplimit:
        print("没有异常值")
    else:
        raise Exception("存在异常值")
```

编写一个简单的一元线性回归模型的函数

```python
from sklearn.linear_model import LinearRegression

def regr_func(x,y):
    regr = LinearRegression()
    x = x.values.reshape(-1,1)
    regr.fit(x,y)
    coef = "%.3f"%regr.coef_
    intercept = "%.3f"%regr.intercept_
    print("y = "+intercept+" + "+"("+coef+" * x"+")")
```

编写一个简单的相关系数函数，并计算这些因素间的相关系数

```python
import numpy as np
def xy_cor(x,y):
    m_x = np.mean(x)
    m_x2 = np.mean(np.square(x))
    m_y = np.mean(y)
    m_y2 = np.mean(np.square(y))
    m_xy = np.mean(x*y)
    v_x = m_x2 - np.square(m_x)
    v_y = m_y2 - np.square(m_y)
    cov_xy = m_xy - m_x*m_y
    cor_xy = cov_xy/(np.sqrt(v_x)*np.sqrt(v_y))
    return cor_xy
    
cnt = df1['cnt'].values
temp = df1['temp'].values
atemp = df1['atemp'].values
hum = df1['hum'].values
windspeed = df1['windspeed'].values
holiday = df1['holiday'].values
weathersit = df1['weathersit'].values

c1 = xy_cor(temp,cnt)
c2 = xy_cor(atemp,cnt)
c3 = xy_cor(hum,cnt)
c4 = xy_cor(windspeed,cnt)
c5 = xy_cor(holiday,cnt)
c6 = xy_cor(weathersit,cnt)
print("cnt与temp的相关系数为: ","%.3f"%c1)
print("cnt与atemp的相关系数为: ","%.3f"%c2)
print("cnt与hum的相关系数为: ","%.3f"%c3)
print("cnt与windspeed的相关系数为: ","%.3f"%c4)
print("cnt与holiday的相关系数为: ","%.3f"%c5)
print("cnt与weathersit的相关系数为: ","%.3f"%c6)
```

定义一个多元线性回归的类

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

class linreg:
    def __init__(self,x,y):
        self.x = x
        self.y = y
        
    def linreg_intercept(self):
        x_train,x_test,y_train,y_test = train_test_split(self.x,self.y,test_size=0.2,random_state=100)
        lr = LinearRegression()
        model = lr.fit(x_train,y_train)
        print("%.3f"%lr.intercept_)
        
    def linreg_coef(self):
        x_train,x_test,y_train,y_test = train_test_split(self.x,self.y,test_size=0.2,random_state=100)
        lr = LinearRegression()
        model = lr.fit(x_train,y_train)
        coef = []
        n = len(lr.coef_)
        for i in range(n):
            c = "%.3f"%lr.coef_[i]
            coef.append(c)
        print(coef)
        
    def assess(self):
        x_train,x_test,y_train,y_test = train_test_split(self.x,self.y,test_size=0.2,random_state=100)
        lr = LinearRegression()
        model = lr.fit(x_train,y_train)
        y_pred = lr.predict(x_test)
        plt.figure()
        plt.plot(range(len(y_pred)),y_pred,'b',label="predict")
        plt.plot(range(len(y_pred)),y_test,'r',label="test")
        plt.legend(loc="upper right")
        plt.show()
        
        sum_mean = 0
        for i in range(len(y_pred)):
            sum_mean += (y_pred[i]-y_test.values[i])**2
        sum_erro = np.sqrt(sum_mean/len(x_test))
        print("RMSE by hand:","%.3f"%sum_erro)
           
```

建立cnt 与 temp,atemp,hum,windspeed,holiday,weathersit的多元线性回归模型

```python
X = df1.loc[:,('temp','atemp','hum','windspeed','holiday','weathersit')]
Y = df1.loc[:,'cnt']
linr = linreg(X,Y)
Inte = linr.linreg_intercept()
coef = linr.linreg_coef()
linr.assess()
```

建立cnt与temp,windspeed,weathersit的多元线性回归模型

```python
X1 = df1.loc[:,('temp','windspeed','weathersit')]
Y1 = df1.loc[:,'cnt']
linr = linreg(X1,Y1)
Inte = linr.linreg_intercept()
coef = linr.linreg_coef()
linr.assess()
```
